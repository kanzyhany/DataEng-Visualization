## **Data Cleaning**

```python
# ---------------------------
# CLEAN COLUMNS
# ---------------------------
def clean_columns(df):
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("/", "_")
    )
    return df

df_crashes = clean_columns(df_crashes)
df_persons = clean_columns(df_persons)
```

```python
# ---------------------------
# MISSING VALUES (DROP + IMPUTE)
# ---------------------------
def clean_missing(df):
    df = df.copy()
    missing_ratio = df.isnull().mean()

    # Drop columns with >60% missing
    cols_to_drop = missing_ratio[missing_ratio > 0.6].index
    df = df.drop(columns=cols_to_drop)

    # Drop rows missing collision_id
    if "collision_id" in df.columns:
        df = df.dropna(subset=["collision_id"])

    # Impute numeric with median
    for col in df.select_dtypes(include=[np.number]).columns:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].median())

    # Impute categorical with "Unknown"
    for col in df.select_dtypes(include=["object"]).columns:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna("Unknown")

    return df

df_crashes = clean_missing(df_crashes)
df_persons = clean_missing(df_persons)
```

```python

# ---------------------------
# OUTLIER HANDLING (IQR, FAST)
# ---------------------------
def fix_outliers(df):
    df = df.copy()

    for col in df.select_dtypes(include=[np.number]).columns:
        if df[col].nunique() <= 1:
            continue

        if "injured" in col or "killed" in col:
            continue  # preserve true values

        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        low = Q1 - 1.5 * IQR
        high = Q3 + 1.5 * IQR

        df[col] = df[col].clip(low, high)

    return df

df_crashes = fix_outliers(df_crashes)
df_persons = fix_outliers(df_persons)
```

```python

# ---------------------------
# REMOVE DUPLICATES (CORRECT FIX)
# ---------------------------
def remove_duplicates(df):
    return df.drop_duplicates().copy()

df_crashes = remove_duplicates(df_crashes)
df_persons = remove_duplicates(df_persons)

print("Cleaning finished safely.")
```

```python
# Safe datetime parsing
df_crashes['crash_datetime'] = pd.to_datetime(
    df_crashes['crash_date'] + ' ' + df_crashes['crash_time'],
    errors='coerce'
)

df_persons['crash_datetime'] = pd.to_datetime(
    df_persons['crash_date'] + ' ' + df_persons['crash_time'],
    errors='coerce'
)
```

```python
# ---------------------------
# SAVE SAFELY
# ---------------------------
df_crashes.to_pickle('crashes_cleaned.pkl')
df_persons.to_pickle('persons_cleaned.pkl')

df_crashes.to_csv('crashes_cleaned.csv', index=False)
df_persons.to_csv('persons_cleaned.csv', index=False)

print("Saved cleaned files.")
```

```python
import pandas as pd
from rapidfuzz import fuzz
from sklearn.cluster import AgglomerativeClustering
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import pairwise_distances


def auto_clean_categories(df, column, similarity_threshold=75):
    values = df[column].dropna().unique()
    values_list = list(values)

    n = len(values_list)
    distance_matrix = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            sim = fuzz.token_sort_ratio(values_list[i], values_list[j])
            distance_matrix[i][j] = 100 - sim

    clustering = AgglomerativeClustering(
        n_clusters=None,
        metric='precomputed',       # ✔ fixed argument
        linkage='average',
        distance_threshold=100 - similarity_threshold
    ).fit(distance_matrix)

    df_map = {}

    freq = df[column].value_counts()

    for cluster_id in set(clustering.labels_):
        cluster_values = [values_list[i] for i in range(n) if clustering.labels_[i] == cluster_id]

        dominant = freq[freq.index.isin(cluster_values)].idxmax()

        for v in cluster_values:
            df_map[v] = dominant

    df[column] = df[column].map(df_map).fillna(df[column])
    return df
```
