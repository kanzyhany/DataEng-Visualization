```python
#crash

# 1️⃣ IQR method for numeric columns
num_cols = df_crashes.select_dtypes(include=[np.number]).columns

for col in num_cols:
    # Skip columns with all zeros or only one unique value
    if df_crashes[col].nunique() <= 1:
        continue

    Q1 = df_crashes[col].quantile(0.25)
    Q3 = df_crashes[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip values outside the IQR range
    df_crashes[col] = df_crashes[col].clip(lower_bound, upper_bound)



# -------------------------------
# 3️⃣ Summary of changes
# -------------------------------
print("✅ Outlier detection and correction complete.")
print("Numeric columns checked:", list(num_cols))
```

```python
#removing duplicates
def remove_duplicates(df_crashes, df_persons):
    """
    Remove duplicate records from both datasets
    """
    print("\n" + "=" * 60)
    print("REMOVING DUPLICATES")
    print("=" * 60)

    crashes_clean = df_crashes.copy()
    persons_clean = df_persons.copy()

    # Store original shapes
    original_crashes_shape = crashes_clean.shape
    original_persons_shape = persons_clean.shape

    # =========================================================================
    # Remove Exact Duplicates
    # =========================================================================
    print("\n1. REMOVING EXACT DUPLICATES")
    print("-" * 30)

    # Remove exact duplicates from crashes dataset
    crashes_before = len(crashes_clean)
    crashes_clean = crashes_clean.drop_duplicates()
    crashes_after = len(crashes_clean)
    crashes_duplicates = crashes_before - crashes_after

    print(f"Crashes dataset:")
    print(f"  Before: {crashes_before:,} records")
    print(f"  After:  {crashes_after:,} records")
    print(f"  Removed: {crashes_duplicates:,} exact duplicates ({crashes_duplicates/crashes_before*100:.2f}%)")

    # Remove exact duplicates from persons dataset
    persons_before = len(persons_clean)
    persons_clean = persons_clean.drop_duplicates()
    persons_after = len(persons_clean)
    persons_duplicates = persons_before - persons_after

    print(f"Persons dataset:")
    print(f"  Before: {persons_before:,} records")
    print(f"  After:  {persons_after:,} records")
    print(f"  Removed: {persons_duplicates:,} exact duplicates ({persons_duplicates/persons_before*100:.2f}%)")

#Summary Statistics
    print("\n3. DUPLICATE REMOVAL SUMMARY")
    print("-" * 30)

    total_crashes_removed = original_crashes_shape[0] - len(crashes_clean)
    total_persons_removed = original_persons_shape[0] - len(persons_clean)

    print(f"Crashes Dataset:")
    print(f"  Original: {original_crashes_shape[0]:,} records")
    print(f"  Final:    {len(crashes_clean):,} records")
    print(f"  Total Removed: {total_crashes_removed:,} records ({total_crashes_removed/original_crashes_shape[0]*100:.2f}%)")

    print(f"Persons Dataset:")
    print(f"  Original: {original_persons_shape[0]:,} records")
    print(f"  Final:    {len(persons_clean):,} records")
    print(f"  Total Removed: {total_persons_removed:,} records ({total_persons_removed/original_persons_shape[0]*100:.2f}%)")

    return crashes_clean, persons_clean

# Execute duplicate removal
print("Starting duplicate removal process...")
crashes_deduped, persons_deduped = remove_duplicates(df_crashes, df_persons)
```

## **Data Integration**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

KEY = "COLLISION_ID"
KEY_ALIASES = ["COLLISION_ID", "collision_id", "CRASH_ID", "ACCIDENT_ID", "COL_ID", "COLLISIONID"]
CRASH_TYPE_COLS = ["CRASH_TYPE", "CRASH_TYPE_DESC", "CRASH_CATEGORY"]
INJURY_TYPE_COLS = ["INJURY_SEVERITY", "PERSON_INJURY", "INJURY_TYPE"]


def _normalize_name(name: str) -> str:
	return "".join(ch for ch in name.strip().upper() if ch.isalnum() or ch == "_")


def _resolve_key_name(df: pd.DataFrame, key: str, aliases: list[str]) -> str:
	# Try exact
	if key in df.columns:
		return key
	# Try case-insensitive / normalized
	norm_to_actual = {_normalize_name(c): c for c in df.columns}
	candidates = [key] + aliases
	for cand in candidates:
		norm = _normalize_name(cand)
		if norm in norm_to_actual:
			return norm_to_actual[norm]
	raise KeyError(f"Join key not found. Looked for {key!r} and aliases {aliases}. Available: {list(df.columns)[:20]}...")


def coalesce_columns(df: pd.DataFrame, preferred_names: list[str], new_name: str) -> pd.DataFrame:
	available = [c for c in preferred_names if c in df.columns]
	if not available:
		return df
	df[new_name] = df[available].bfill(axis=1).ffill(axis=1).iloc[:, 0]
	keep = [c for c in df.columns if c not in available or c == new_name]
	return df[keep]


def ensure_key_dtype_alignment(left: pd.DataFrame, right: pd.DataFrame, key: str) -> tuple[pd.DataFrame, pd.DataFrame]:
	left = left.copy()
	right = right.copy()
	left_key_col = _resolve_key_name(left, key, KEY_ALIASES)
	right_key_col = _resolve_key_name(right, key, KEY_ALIASES)

	def to_key_str(s: pd.Series) -> pd.Series:
		s = s.astype("string").str.strip()
		s = s.str.replace(r"\.0$", "", regex=True)
		return s

	# Create a standardized KEY column even if the source column has a different name
	left[KEY] = to_key_str(left[left_key_col])
	right[KEY] = to_key_str(right[right_key_col])
	return left, right


def standardize_columns(df: pd.DataFrame) -> pd.DataFrame:
	rename_map = {c: c.strip().upper() for c in df.columns}
	df = df.rename(columns=rename_map)
	df = coalesce_columns(df, CRASH_TYPE_COLS, "CRASH_TYPE_STD")
	df = coalesce_columns(df, INJURY_TYPE_COLS, "INJURY_TYPE_STD")
	return df


def normalize_categories(df: pd.DataFrame, col: str, mapping: dict[str, str] | None = None) -> pd.DataFrame:
	if col not in df.columns:
		return df
	s = df[col].astype("string").str.strip().str.upper()
	s = s.replace({
		"KILLED": "FATAL",
		"FATAL INJURY": "FATAL",
		"SERIOUS": "SERIOUS INJURY",
		"MINOR": "MINOR INJURY",
		"NO INJURY": "NO APPARENT INJURY",
		"NONE": "NO APPARENT INJURY",
	})
	if mapping:
		s = s.replace({k.upper(): v for k, v in mapping.items()})
	df[col] = s
	return df


def validate_join_keys(df_crashes: pd.DataFrame, df_persons: pd.DataFrame, key: str):
	if key not in df_crashes.columns:
		left_key_col = _resolve_key_name(df_crashes, key, KEY_ALIASES)
		crash_keys = df_crashes[left_key_col].dropna().astype("string").unique()
	else:
		crash_keys = df_crashes[key].dropna().astype("string").unique()
	if key not in df_persons.columns:
		right_key_col = _resolve_key_name(df_persons, key, KEY_ALIASES)
		person_keys = df_persons[right_key_col].dropna().astype("string").unique()
	else:
		person_keys = df_persons[key].dropna().astype("string").unique()
	inter = np.intersect1d(crash_keys, person_keys)
	print(f"Crash unique keys: {len(crash_keys):,}")
	print(f"Person unique keys: {len(person_keys):,}")
	print(f"Intersecting keys: {len(inter):,}")


def report_counts(df_crashes: pd.DataFrame, df_persons: pd.DataFrame, merged: pd.DataFrame, key: str, how: str):
	print(f"Crashes rows: {len(df_crashes):,}")
	print(f"Persons rows: {len(df_persons):,}")
	print(f"Merged rows ({how}): {len(merged):,}")
	left_keys = merged[key].dropna().astype("string").unique()
	# Resolve source key columns if KEY is not present in sources
	left_key_col = key if key in df_crashes.columns else _resolve_key_name(df_crashes, key, KEY_ALIASES)
	right_key_col = key if key in df_persons.columns else _resolve_key_name(df_persons, key, KEY_ALIASES)
	unmatched_crash = df_crashes[~df_crashes[left_key_col].astype("string").isin(left_keys)]
	unmatched_person = df_persons[~df_persons[right_key_col].astype("string").isin(left_keys)]
	print(f"Unmatched crash keys: {unmatched_crash[left_key_col].astype('string').nunique():,}")
	print(f"Unmatched person keys: {unmatched_person[right_key_col].astype('string').nunique():,}")
```

```python
# Configure join type and run the merge safely
# how options: "inner", "left", "right", "outer"
how = "inner"

# Preflight checks and robust merge
try:
	# Check input dataframes exist
	assert 'df_crashes' in globals(), "df_crashes is not defined above this cell"
	assert 'df_persons' in globals(), "df_persons is not defined above this cell"

	# 1) Align key dtypes and create standardized KEY column
	left_aligned, right_aligned = ensure_key_dtype_alignment(df_crashes, df_persons, KEY)

	# 2) Standardize and normalize
	left_std = standardize_columns(left_aligned)
	right_std = standardize_columns(right_aligned)
	left_std = normalize_categories(left_std, "CRASH_TYPE_STD")
	right_std = normalize_categories(right_std, "INJURY_TYPE_STD")

	# 2b) Re-assert a single KEY column using robust extractor and drop duplicate labels
	if '_extract_key_series' in globals():
		extractor = _extract_key_series
	else:
		def extractor(df, key, aliases):
			norm_targets = {key} | set(aliases)
			upper_targets = {t.upper() for t in norm_targets}
			matches = [c for c in df.columns if c.upper() in upper_targets]
			if not matches:
				raise KeyError(f"Join key not found for {key!r}.")
			sub = df[matches]
			if isinstance(sub, pd.DataFrame):
				coalesced = sub.astype("string").apply(lambda col: col.str.strip().str.replace(r"\.0$", "", regex=True))
				return coalesced.bfill(axis=1).ffill(axis=1).iloc[:, 0]
			return sub.astype("string").str.strip().str.replace(r"\.0$", "", regex=True)
	left_std[KEY] = extractor(left_std, KEY, KEY_ALIASES)
	right_std[KEY] = extractor(right_std, KEY, KEY_ALIASES)
	left_std = left_std.loc[:, ~left_std.columns.duplicated()]
	right_std = right_std.loc[:, ~right_std.columns.duplicated()]

	# 3) Merge with suffixes to avoid collisions
	merged = left_std.merge(right_std, on=KEY, how=how, suffixes=("_CRASH", "_PERSON"))

	print(f"Merged shape: {merged.shape}")
	display(merged.head())
except AssertionError as e:
	print(f"Preflight failed: {e}")
	raise
except KeyError as e:
	print("Key error during merge. Available columns (first 20) ->")
	print("Crashes:", list(df_crashes.columns)[:20] if 'df_crashes' in globals() else None)
	print("Persons:", list(df_persons.columns)[:20] if 'df_persons' in globals() else None)
	raise
except Exception as e:
	print("Unexpected error during merge:", type(e).__name__, str(e))
	raise
```

```python
# Validate integration results safely
try:
	print("=== Pre-join key validation ===")
	validate_join_keys(df_crashes, df_persons, KEY)

	print("\n=== Post-join record validation ===")
	if 'merged' not in globals():
		raise NameError("'merged' is not defined. Run the merge cell above first.")
	report_counts(df_crashes, df_persons, merged, KEY, how)

	# Optional integrity checks for 1-to-many expectation (persons per crash)
	persons_per_collision = merged.groupby(KEY).size()
	print(f"Persons per collision — min: {persons_per_collision.min()}, max: {persons_per_collision.max()}, mean: {persons_per_collision.mean():.2f}")
except Exception as e:
	print("Validation error:", type(e).__name__, str(e))
	raise
```

```python
# Validation plots
# 1) Crash type vs Injury type heatmap (row-normalized %)
crash_col = "CRASH_TYPE_STD" if "CRASH_TYPE_STD" in merged.columns else ( [c for c in CRASH_TYPE_COLS if c in merged.columns] or [None] )[0]
inj_col = "INJURY_TYPE_STD" if "INJURY_TYPE_STD" in merged.columns else ( [c for c in INJURY_TYPE_COLS if c in merged.columns] or [None] )[0]

if crash_col is None or inj_col is None:
	print("Skipping heatmap: crash/injury columns not found")
else:
	ct = pd.crosstab(merged[crash_col], merged[inj_col], dropna=False)
	ct_pct = ct.div(ct.sum(axis=1).replace(0, np.nan), axis=0) * 100
	plt.figure(figsize=(12, 6))
	sns.heatmap(ct_pct, annot=True, fmt=".1f", cmap="Blues", cbar_kws={"label": "% within crash type"})
	plt.title("Crash Type vs Injury Type (% within crash type)")
	plt.xlabel("Injury Type")
	plt.ylabel("Crash Type")
	plt.tight_layout()
	plt.show()

# 2) Distribution of persons per collision
persons_per_crash = merged.groupby(KEY).size().clip(upper=25)
plt.figure(figsize=(10, 4))
sns.histplot(persons_per_crash, bins=25, kde=False)
plt.title("Distribution of persons per collision (clipped at 25)")
plt.xlabel("# persons per collision")
plt.ylabel("Count of collisions")
plt.tight_layout()
plt.show()
```
