## **Data Cleaning**

***clean column names// drop missing values***

```python
#crashh
num_columns_before_crashes = df_crashes.shape[1]
df_crashes.columns = (
    df_crashes.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('/', '_')
)

# Handle missing values (Drop vs Impute)
missing_ratio = df_crashes.isnull().mean()
cols_to_drop = missing_ratio[missing_ratio > 0.6].index
df_crashes.drop(columns=cols_to_drop, inplace=True)

if 'collision_id' in df_crashes.columns:
    df_crashes.dropna(subset=['collision_id'], inplace=True)

# Impute numeric columns
num_cols = df_crashes.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    if df_crashes[col].isnull().sum() > 0:
        df_crashes[col] = df_crashes[col].fillna(df_crashes[col].median()) #replaces na with median

# Impute categorical columns
cat_cols = df_crashes.select_dtypes(include=['object']).columns
for col in cat_cols:
    if df_crashes[col].isnull().sum() > 0:
        df_crashes[col] = df_crashes[col].fillna('Unknown')

num_columns_after_crashes = df_crashes.shape[1]


# Show summary
print(f"Dropped columns: {list(cols_to_drop)} because missing_ratio > 0.6")
print(f"Number of coulmns before dropping: {num_columns_before_crashes } \n Number of coulmns after dropping: {num_columns_after_crashes}")
print("Remaining missing values per column:")
print(df_crashes.isnull().sum().sort_values(ascending=False).head(10))
```

```python
#persons
num_columns_before_persons = df_persons.shape[1]
df_persons.columns = (
    df_persons.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('/', '_')
)

# Handle missing values (Drop vs Impute)
missing_ratio = df_persons.isnull().mean()
cols_to_drop = missing_ratio[missing_ratio > 0.6].index
df_persons.drop(columns=cols_to_drop, inplace=True)

if 'collision_id' in df_persons.columns:
    df_persons.dropna(subset=['collision_id'], inplace=True)

# Impute numeric columns
num_cols = df_persons.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    if df_persons[col].isnull().sum() > 0:
        df_persons[col] = df_persons[col].fillna(df_persons[col].median()) #replaces na with median

# Impute categorical columns
cat_cols = df_persons.select_dtypes(include=['object']).columns
for col in cat_cols:
    if df_persons[col].isnull().sum() > 0:
        df_persons[col] = df_persons[col].fillna('Unknown')

num_columns_after_persons = df_persons.shape[1]


# Show summary
print(f"Dropped columns: {list(cols_to_drop)} because missing_ratio > 0.6")
print(f"Number of coulmns before dropping: {num_columns_before_persons } \n Number of coulmns after dropping: {num_columns_after_persons}")
print("Remaining missing values per column:")
print(df_persons.isnull().sum().sort_values(ascending=False).head(10))
```

***ourliers***

```python
#crash

# 1️⃣ IQR method for numeric columns
num_cols = df_crashes.select_dtypes(include=[np.number]).columns

for col in num_cols:
    # Skip columns with all zeros or only one unique value
    if df_crashes[col].nunique() <= 1:
        continue

    Q1 = df_crashes[col].quantile(0.25)
    Q3 = df_crashes[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip values outside the IQR range
    df_crashes[col] = df_crashes[col].clip(lower_bound, upper_bound)



# -------------------------------
# 3️⃣ Summary of changes
# -------------------------------
print("✅ Outlier detection and correction complete.")
print("Numeric columns checked:", list(num_cols))
```

```python
#persons

# 1️⃣ IQR method for numeric columns
num_cols = df_persons.select_dtypes(include=[np.number]).columns

for col in num_cols:
    # Skip columns with all zeros or only one unique value
    if df_persons[col].nunique() <= 1:
        continue

    Q1 = df_persons[col].quantile(0.25)
    Q3 = df_persons[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip values outside the IQR range
    df_persons[col] = df_persons[col].clip(lower_bound, upper_bound)



# -------------------------------
# 3️⃣ Summary of changes
# -------------------------------
print("✅ Outlier detection and correction complete.")
print("Numeric columns checked:", list(num_cols))
```

***removing duplicates***

```python
#removing duplicates
def remove_duplicates(df_crashes, df_persons):
    """
    Remove duplicate records from both datasets
    """
    print("\n" + "=" * 60)
    print("REMOVING DUPLICATES")
    print("=" * 60)

    crashes_clean = df_crashes.copy()
    persons_clean = df_persons.copy()

    # Store original shapes
    original_crashes_shape = crashes_clean.shape
    original_persons_shape = persons_clean.shape

    # =========================================================================
    # Remove Exact Duplicates
    # =========================================================================
    print("\n1. REMOVING EXACT DUPLICATES")
    print("-" * 30)

    # Remove exact duplicates from crashes dataset
    crashes_before = len(crashes_clean)
    crashes_clean = crashes_clean.drop_duplicates()
    crashes_after = len(crashes_clean)
    crashes_duplicates = crashes_before - crashes_after

    print(f"Crashes dataset:")
    print(f"  Before: {crashes_before:,} records")
    print(f"  After:  {crashes_after:,} records")
    print(f"  Removed: {crashes_duplicates:,} exact duplicates ({crashes_duplicates/crashes_before*100:.2f}%)")

    # Remove exact duplicates from persons dataset
    persons_before = len(persons_clean)
    persons_clean = persons_clean.drop_duplicates()
    persons_after = len(persons_clean)
    persons_duplicates = persons_before - persons_after

    print(f"Persons dataset:")
    print(f"  Before: {persons_before:,} records")
    print(f"  After:  {persons_after:,} records")
    print(f"  Removed: {persons_duplicates:,} exact duplicates ({persons_duplicates/persons_before*100:.2f}%)")

#Summary Statistics
    print("\n3. DUPLICATE REMOVAL SUMMARY")
    print("-" * 30)

    total_crashes_removed = original_crashes_shape[0] - len(crashes_clean)
    total_persons_removed = original_persons_shape[0] - len(persons_clean)

    print(f"Crashes Dataset:")
    print(f"  Original: {original_crashes_shape[0]:,} records")
    print(f"  Final:    {len(crashes_clean):,} records")
    print(f"  Total Removed: {total_crashes_removed:,} records ({total_crashes_removed/original_crashes_shape[0]*100:.2f}%)")

    print(f"Persons Dataset:")
    print(f"  Original: {original_persons_shape[0]:,} records")
    print(f"  Final:    {len(persons_clean):,} records")
    print(f"  Total Removed: {total_persons_removed:,} records ({total_persons_removed/original_persons_shape[0]*100:.2f}%)")

    return crashes_clean, persons_clean

# Execute duplicate removal
print("Starting duplicate removal process...")
crashes_deduped, persons_deduped = remove_duplicates(df_crashes, df_persons)
```

***standardizing date/time***

```python
#crash

df_crashes['crash_date'] = pd.to_datetime(df_crashes['crash_date'], errors='coerce')
df_crashes['crash_time'] = pd.to_datetime(df_crashes['crash_time'], format='%H:%M', errors='coerce')


df_crashes['crash_datetime'] = df_crashes.apply(
    lambda row: datetime.combine(
        row['crash_date'].date() if pd.notnull(row['crash_date']) else datetime.min.date(),
        row['crash_time'].time() if pd.notnull(row['crash_time']) else datetime.min.time()
    ),
    axis=1
)

print(df_crashes[['crash_date', 'crash_time', 'crash_datetime']].head())
```

```python
#person

df_persons['crash_date'] = pd.to_datetime(df_persons['crash_date'], errors='coerce')
df_persons['crash_time'] = pd.to_datetime(df_persons['crash_time'], format='%H:%M', errors='coerce')


df_persons['crash_datetime'] = df_persons.apply(
    lambda row: datetime.combine(
        row['crash_date'].date() if pd.notnull(row['crash_date']) else datetime.min.date(),
        row['crash_time'].time() if pd.notnull(row['crash_time']) else datetime.min.time()
    ),
    axis=1
)

print(df_persons[['crash_date', 'crash_time', 'crash_datetime']].head())
```

***standardizing categorical columns***

```python
#crashes
string_cols = df_crashes.select_dtypes(include='object').columns

# 1. Strip spaces, unify casing (title case)
for col in string_cols:
    df_crashes[col] = df_crashes[col].astype(str).str.strip().str.title()

# 2. Replace inconsistent or missing categorical values
df_crashes.replace(
    {
        'Nan': None,
        'N/A': 'Unknown',
        'Na': 'Unknown',
        'NaN': 'Unknown',
        'Unspecified': 'Unknown',
        '': 'Unknown'
    },
    inplace=True
)

# 3. Standardize borough names (if any inconsistencies exist)
df_crashes['borough'] = df_crashes['borough'].replace({
    'Ny': 'Manhattan',
    'Nyc': 'Manhattan',
    'Bklyn': 'Brooklyn',
    'Bronx County': 'Bronx'
})

#making sure there are no unknowns
df_crashes['borough'].unique()
df_crashes['contributing_factor_vehicle_1'].unique()[:10]

```

```python
# persons
string_cols_2 = df_persons.select_dtypes(include='object').columns

# 1. Strip spaces, unify casing (lowercase all person-related text fields)
for col in string_cols_2:
    df_persons[col] = df_persons[col].astype("string").str.strip().str.lower()

# 2. Replace inconsistent or missing categorical values with normalized tokens (lowercase)
df_persons.replace(
    {
        'nan': None,
        'n/a': 'unknown',
        'na': 'unknown',
        'NaN': 'unknown',
        'unspecified': 'unknown',
        '': 'unknown'
    },
    inplace=True
)

# 3. Optional: standardize common fields explicitly if present
if 'person_sex' in df_persons.columns:
    df_persons['person_sex'] = df_persons['person_sex'].replace({
        'm': 'male',
        'f': 'female',
        'u': 'unknown',
        'x': 'unknown'
    })
```
